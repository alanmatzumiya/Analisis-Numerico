```python
a = 50
b = 52

print bin(10), bin(5)
```

```python
c = ~ a

print c, bin(c)
```

```python
c = a & b

print c, bin(c)
```

```python
c = a | b

print c, bin(c)
```

```python
c = a ^ b

print c, bin(c)
```

```python
c = a << 3

print c, bin(c)
```

```python
c = b >> 4

print c, bin(c)
```
